Επιλέγουμε την στήλη renewables_electricity η οποία περιγράφει την γενική παραγωγή ηλεκτρισμού από ανανεώσιμες πηγές ενέργειας για κάθε μία χώρα. Όμως η στήλη αυτή στο country == Europe απεικονίζει το άθροισμα όλων των υπόλοιπων παραγωγών. Οπότε εμείς απομονώνουμε και εξαιρούμε αυτή τη στήλη και στην συυνέχεια την διαιρούμε με τον αριθμό των χωρών που ανήκουν στην Ευρωπαϊκή Ένωση. Έτσι υπολογίζουμε τον μ.ο. της παραγωγής, τον απεικονίζουμε πιο έντονα (μεγαλύτερο πάχος) από τις υπόλοιπες χώρες και δημιουργούμε ένα διάγραμμα που διαπιστώνεται η αυξητική τάση των ανανεώσημων πηγών ενέργειας την επιλεγμένη δεκαετία.

```python
import matplotlib.pyplot as plt
import numpy as np


r_elec_eu = data_2012_2022[data_2012_2022['country'] == 'Europe']
r_elec_eu['renewables_electricity'] /= 27
r_elec = data_2012_2022[data_2012_2022['country'] != 'Europe']
r_elec_other = r_elec.groupby('country')

plt.figure(figsize=(12, 8))
plt.plot(r_elec_eu['year'], r_elec_eu['renewables_electricity'], label='Europe Average',linewidth = 5)

for country, data in r_elec_other:
    plt.plot(data['year'], data['renewables_electricity'], label=country)

plt.xlabel('Year')
plt.ylabel('Terawatt-Hours')
plt.title('Renewable Electricity Generation For Year By Each EU Country (2012-2022)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Παρακάτω παρουσιάζουμε την σύγκριση μεταξύ της γενικής παραγωγής ηλεκτρικής ενέργειας στην Ευρώπη και της παραγωγής ηλεκτρικής ενέργειας από τις ανανεώσιμες πηγές ενέργειας. Διαπιστώνουμε ότι στο διάγραμμα η παραγωγή ηλεκτρικής ενέργειας από ανανεώσιμες πηγές ενέργειας φαίνεται σαν να μένει σταθερή και να είναι μία ευθεία. Παρόλο που στο παραπάνω διάγραμμα φαίνεται η αυξητική τάση της παραγωγής ηλεκτρικής ενέργειας από ανανεώσιμες πηγές σε αυτό το διάγραμμα δεν διαπιστώνεται αυτό. Συμπαιρένουμε δηλαδή ότι η αυξητική τάση παραγωγής ηλεκτρικής ενέργειας από ανανεώσιμες πηγές είναι πολύ μικρή σε σχέση με την συνολική παραγωγή ηλεκτρικής ενέργειας. 

```python
plt.figure(figsize=(12, 8))
plt.plot(r_elec_eu['year'], r_elec_eu['renewables_electricity'], label='Renewable Electricity Generation', linewidth=3)
plt.plot(r_elec_eu['year'], r_elec_eu['electricity_generation'], label='Overall Electricity Generation', linewidth=3)

plt.xlabel('Year')
plt.ylabel('Terawatt-Hours')
plt.title('Difference Between Overall And Renewable Electricity Generation In Europe (2012-2022)')
plt.legend()
plt.show()
```

Στα δύο διαγράμματα που ακολουθούν γίνεται πιο σαφές το πόσο έχουν αυξηθεί τα ποσοστά παραγωγής ηλεκτρικής ενέργειας από ανανεώσιμες πηγές από το 2012, το 2022. Συκρίνουμε δηλαδή την συμμετοχή όλων των ποσοστών παραγωγής ηλεκτρικής ενέργειας στην αρχή της δεκαετίας (2012) με το τέλος της επιλεγμένης δεκαετίας (2022). Η συμμετοχή της ενέργειας από ανανεώσιμες πηγές είναι εμφανές ότι έχει αυξηθεί στο τέλος της δεκαετίας (2022).

```python
data_2012 = r_elec[r_elec['year'] == 2012]

r_elec_2012 = data_2012['renewables_electricity'].values
elec_2012 = data_2012['electricity_generation'].values
countries = data_2012['country'].values

positions = np.arange(len(countries))
width = 0.35

plt.figure(figsize=(12, 8))
plt.bar(positions - width/2, r_elec_2012, width, label='Renewable Electricity', color='blue')
plt.bar(positions + width/2, elec_2012, width, label='Conventional Electricity', color='green')

plt.xlabel('Country')
plt.ylabel('Terrawatts-Hours')
plt.title('Renewable Electricity vs Conventional Electricity Generation in 2012')
plt.xticks(positions, countries, rotation='vertical')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
data_2022 = r_elec[r_elec['year'] == 2022]

r_elec_2022 = data_2022['renewables_electricity'].values
elec_2022 = data_2022['electricity_generation'].values
countries = data_2022['country'].values

positions = np.arange(len(countries))
width = 0.35

plt.figure(figsize=(12, 8))
plt.bar(positions - width/2, r_elec_2022, width, label='Renewable Electricity', color='blue')
plt.bar(positions + width/2, elec_2022, width, label='Conventional Electricity', color='green')

plt.xlabel('Country')
plt.ylabel('Terrawatts-Hours')
plt.title('Renewable Electricity vs Conventional Electricity Generation in 2022')
plt.xticks(positions, countries, rotation='vertical')
plt.legend()

plt.tight_layout()
plt.show()
```

Παρακάτω οπτικοποιούμε τις διαφορετικές πηγές ανανεώσιμης ηλεκτρικής ενέργειας καθώς και το ποσοστό που τους αναλογεί για το χρόνο 2022. Παρατηρούμε πως η υδροηλεκτρική και αιολική ενέργεια είναι μεγαλύτερη της ηλιακής.

```python
columns = ['solar_electricity', 'wind_electricity', 'hydro_electricity', 'biofuel_electricity', 'other_renewable_electricity']
all_renewables_2022 = r_elec.loc[r_elec['year'] == 2022, columns].sum()

plt.figure(figsize=(8, 8))
all_renewables_2022.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Renewable Electricity Generation by Source in 2022')
plt.show()
```

Στο παρακάτω γράφημα χρησιμοποίησα τη βοήθεια του ChatGpt προκειμένου να φτιάξω μία απεικόνιση στο folium που δείχνει το ποσοστό παραγωγής ενέργειας απο ανανεωσιμες πηγές. Κειμένεται απο 12 έως 86 τοις εκατό άν και θα έπρεπε να είναι η κλίμακα από 0 έως 1. Οι χώρες που απεικονίζονται με χρώμα κοντά στο μπλέ έχουν μεγαλύτερο ποσοστό παραγωγής ηλεκτρισμού απο ανανεώσιμες πηγές ενέργειας σε σχέση με τη συνολική παραγωγή τους.

```python
import folium


world_geojson_url = "https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/world-countries.json"

choropleth_data =r_elec[['country', 'renewables_share_elec']]

m = folium.Map(location=[51.1657, 10.4515], zoom_start=3)

folium.Choropleth(
    geo_data=world_geojson_url,
    name='choropleth',
    data=choropleth_data,
    columns=['country', 'renewables_share_elec'],
    key_on='feature.properties.name',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Percentage of Renewable Energy Generation by Country',
).add_to(m)

folium.LayerControl().add_to(m)

m


```
